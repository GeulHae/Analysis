```python
rv = norm(loc = 9.5326087, scale = 2.55510)                                      # 평균 9.184783, 표준편차 2.877556인 정규분포
x = np.arange(0, 19.2, 0.01)                                                    # X 확률변수 범위
y = rv.pdf(x)                                                                    # X 범위에 따른 정규확률밀도값

fig, ax = plt.subplots(1,1)
ax.plot(x, y, ms=8, label = 'normal pdf')
ax.vlines(x, 0, y, colors='lightsteelblue', lw =5, alpha =0.3)
ax.set_ylim([0,0.2])                                                             # y축 범위

plt.gca().axes.yaxis.set_visible(False)                                          # y축 범위 삭제
plt.axvline(9.184783, 0, 0.8, color='lightgray', linestyle='--', linewidth=2)    # 평균 세로점선
plt.axvline(9.9, 0, 0.9, color='darkred', linestyle='solid', linewidth=1)        # 중앙값 세로실선
```
